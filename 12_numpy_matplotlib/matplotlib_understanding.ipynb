comparision of 2 graphs in matplotlib

```python
def stock_price_calc(x,m=4,c=10):
    return m*x+c

x=[i for i in range(1,13) ]

y_2=[stock_price_calc(i,m=-3,c=5) for i in x]
y_3=[stock_price_calc(i,m=-6,c=1) for i in x]

plt.plot(x,y_2)

plt.plot(x,y_3)
plt.legend('2025','2023')
plt.xlabel('Months')
plt.ylabel('Stock price')

```
